```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
import pandas as pa
import math

def function(x):
    return np.log(x+1)

def f(x,y):
    return np.log(x+1)*(y-1)

def g(x):
    return -np.log(x+1)*np.exp((-x-1)*np.log(x+1)+x)
```

```python
t = np.linspace(0, 50, 1000)
plt.plot(t,g(t))
```

# Euler

```python
def euler(y0, f, n):
    h=n[1]-n[0]
    y = []
    for i in n:
        y.append(y0)
        y0 = f(i,y0)*h + y0
    return np.array(y)
```

```python
euler1 =euler(2,f,t)
plt.plot(t,euler1)
plt.plot(t,g(t),'r')
```

# Runge-Kutta 4th order

```python
def runge_kutta(f, x0, y0, h, n):
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    x[0] = x0
    y[0] = y0

    for i in range(n):
        k1 = h * f(x[i], y[i])
        k2 = h * f(x[i] + h/2, y[i] + k1/2)
        k3 = h * f(x[i] + h/2, y[i] + k2/2)
        k4 = h * f(x[i] + h, y[i] + k3)

        y[i+1] = y[i] + (k1 + 2*k2 + 2*k3 + k4) / 6
        x[i+1] = x[i] + h

    return x, y
```

```python
# Solution 1
x0_1 = 0.1
y0_1 = 0.2
h_1 = 0.1
n_1 = 100

x1, y1 = runge_kutta(f, x0_1, y0_1, h_1, n_1)

# Solution 2
x0_2 = 0.1
y0_2 = 1.0
h_2 = 0.1
n_2 = 100

x2, y2 = runge_kutta(f, x0_2, y0_2, h_2, n_2)

# Solution 3
x0_3 = 0.1
y0_3 = 5.0
h_3 = 0.1
n_3 = 100

x3, y3 = runge_kutta(f, x0_3, y0_3, h_3, n_3)

# Plot the solutions
plt.plot(x1, y1, label='y0 = 0.2')
plt.plot(x2, y2, label='y0 = 1.0')
plt.plot(x3, y3, label='y0 = 5.0')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
